### display of the states, actions and rewards of the Gamblerâ€™s MDP problem with 500 states

```python
class GamblersProblem:
    def __init__(self, goal=500):
        self.goal = goal  # Target amount of money to win
        self.states = list(range(goal + 1))  # States are from 0 to 500 (inclusive)

    def actions(self, state):
        # Can bet any amount from 1 to min(state, goal - state)
        return list(range(1, min(state, self.goal - state) + 1))

    def reward(self, state):
        # Reward of 1 if goal is reached, otherwise 0
        if state == self.goal:
            return 1
        else:
            return 0

    def display(self):
        # Print states, possible actions from each state, and rewards
        for state in self.states:
            possible_actions = self.actions(state)
            reward_for_state = self.reward(state)
            print(f"State: {state}, Actions: {possible_actions}, Reward: {reward_for_state}")

# Create an instance of the problem
gamblers_problem = GamblersProblem()

# Display states, actions, and rewards
gamblers_problem.display()
```
